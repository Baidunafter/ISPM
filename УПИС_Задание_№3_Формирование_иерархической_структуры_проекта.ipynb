# 1. ТЕОРЕТИЧЕСКИЕ ОСНОВЫ И ОБЗОР ПРЕДМЕТНОЙ ОБЛАСТИ

## 1.1. Современное состояние обработки данных дистанционного зондирования Земли

### 1.1.1. Задачи мониторинга территорий по спутниковым данным Sentinel-2
Рассматриваются основные прикладные задачи, решаемые с помощью данных Sentinel-2: классификация земного покрова, мониторинг сельскохозяйственных угодий, оценка последствий чрезвычайных ситуаций, контроль застройки. Анализируются требования к пространственному и временному разрешению данных.

### 1.1.2. Традиционные методы обработки спутниковых снимков и их ограничения
Описываются классические подходы к сегментации и классификации: пороговые методы, объектно-ориентированный анализ, методы на основе спектральных индексов (NDVI, NDWI). Выявляются их недостатки: трудоёмкость ручной настройки, низкая адаптивность к региональным особенностям, ограниченная точность на гетерогенных ландшафтах.

### 1.1.3. Переход к методам глубокого обучения в задачах ДЗЗ
Рассматривается эволюция применения нейронных сетей для анализа спутниковых данных: от первых свёрточных архитектур (FCN, U-Net) до современных гибридных моделей. Обосновывается актуальность использования глубокого обучения для автоматизации мониторинга территорий.

## 1.2. Обзор современных нейросетевых архитектур для сегментации данных ДЗЗ

### 1.2.1. Свёрточные нейронные сети (CNN) и их развитие в задачах ДЗЗ
Анализируются классические CNN-архитектуры для сегментации: U-Net, DeepLabV3+, HRNet. Рассматриваются их модификации для работы с мультиспектральными данными и большими пространственными разрешениями. Оцениваются преимущества и ограничения чисто свёрточных подходов.

### 1.2.2. Гибридные архитектуры: CNN-Transformer и CNN-Mamba
Детально разбираются принципы работы гибридных архитектур. Описываются модели семейства CNN-Transformer (SegFormer, CFFormer) и CNN-Mamba (MF-Mamba, Vision Mamba). Анализируются механизмы совмещения локальных признаков (CNN) и глобального контекста (Self-Attention, State Space Models).

### 1.2.3. Фундаментальные модели и подходы JEPA для данных ДЗЗ
Рассматриваются современные фундаментальные модели (AnySat, TerraMind, Prithvi), предобученные на больших объёмах геопространственных данных. Описывается архитектура JEPA (Joint Embedding Predictive Architecture) и её преимущества для задач с ограниченной разметкой.

### 1.3.1. Коммерческие облачные платформы (ArcGIS, Picterra, Mapflow)
Проводится функциональный и технологический анализ коммерческих решений. Выявляются их сильные стороны (удобство, готовые модели) и ограничения (закрытость архитектуры, невозможность сравнения нескольких моделей, высокая стоимость).

Описывается создание веб-интерфейса: страница с картой Leaflet, панель управления (выбор области интереса, выбор модели, запуск анализа), отображение результатов в виде векторных слоёв, реализация режима сравнения двух моделей (side-by-side или переключение слоёв).

### 3.2.3. Интеграция дообученных моделей в систему
Рассматривается процесс упаковки обученных моделей (сохранение весов, создание унифицированного интерфейса загрузки), обеспечение корректной предобработки входных снимков и постобработки выходных масок (конвертация в GeoJSON).

## 3.3. Тестирование, оценка эффективности и перспективы развития

### 3.3.1. Функциональное тестирование системы
Проводится проверка работоспособности всех заявленных функций: загрузка снимка, запуск каждой из трёх моделей, корректность отображения результатов, работа режима сравнения, экспорт данных.

### 3.3.2. Оценка масштабируемости и производительности
Выполняется нагрузочное тестирование (или его имитация): измеряется время обработки снимков разного размера, оценивается эффективность тайлинга, проверяется возможность параллельной обработки при добавлении дополнительных воркеров.

### 3.3.3. Направления дальнейшего развития и рекомендации по внедрению
Формулируются предложения по расширению функционала (добавление новых моделей, поддержка других источников данных, реализация пакетной обработки временных серий). Даются рекомендации по практическому применению системы в организациях, занимающихся мониторингом территорий.
