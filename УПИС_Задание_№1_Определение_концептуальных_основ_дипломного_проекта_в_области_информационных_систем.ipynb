#### **1. Обоснование необходимости разработки**


Для кого предназначено разрабатываемое Вами решение и какие потребности целевой аудитории оно призвано удовлетворить?




*Обоснуйте актуальность решаемой проблемы для конкретных заинтересованных сторон.*

Разрабатываемая информационная система предназначена для **специалистов в области геоинформатики, экологического мониторинга, сельского и городского планирования**, а также **научных сотрудников**, анализирующих спутниковые данные. Основные целевые группы и их потребности:

| Целевая аудитория | Потребности |
| :--- | :--- |
| **ГИС-аналитики, сотрудники региональных администраций** | Оперативное получение карт застройки, водных объектов и типов землепользования без ручной оцифровки |
| **Экологи, специалисты МЧС** | Быстрая оценка площади затоплений, вырубок и других изменений ландшафта |
| **Агрономы, сельхозпроизводители** | Мониторинг состояния посевов, оценка площадей пашни и пастбищ |
| **Исследователи в области ДЗЗ** | Инструмент для объективного сравнения современных нейросетевых архитектур |

**Актуальность решаемой проблемы** обусловлена следующими факторами:

1. **Высокая стоимость ручной обработки.** Оцифровка спутникового снимка для территории среднего района (~5 000 км²) занимает у специалиста 2–4 недели. При средней зарплате ГИС-специалиста в России около 100 000 руб./мес. и стоимости услуг векторизации от 10 000 руб. за проект ручной анализ даже одного района обходится дорого, а для области (100 000+ км²) становится экономически нецелесообразным. Автоматизация сокращает время до минут и на порядок снижает затраты.

2. **Необходимость адаптации к локальной специфике.** Модели, обученные на глобальных датасетах (Европа, США), дают существенное падение точности на территории России из-за особенностей ландшафта (тундра, тайга, мерзлота, болота) и фенологии. Требуется инструмент, позволяющий дообучить и сравнить несколько архитектур именно на региональных данных.

3. **Разрыв между наукой и практикой.** Современные нейросетевые модели существуют в виде исследовательского кода на GitHub. Их использование требует навыков Python, настройки GPU-окружения и работы с геоформатами. Целевая аудитория (ГИС-специалисты, кадастровые инженеры, агрономы) такими компетенциями не обладает, из-за чего передовые методы остаются невостребованными.

4. **Отсутствие удобного инструмента сравнения.** Специалисту важно понимать, какая модель даёт лучший баланс точности и скорости для его задачи. Академические бенчмарки не позволяют загружать свои данные, коммерческие платформы имеют закрытую архитектуру и не дают сравнивать альтернативы. Веб-сервисы работают как «чёрный ящик». Не хватает открытой системы, объединяющей загрузку своих данных, выбор из нескольких моделей и их прямое сравнение в едином интерфейсе.

Таким образом, разрабатываемая система решает проблему **доступности и адаптируемости передовых методов глубокого обучения для прикладных специалистов**, предоставляя инструмент для **обоснованного выбора нейросетевой архитектуры** под конкретную задачу и регион.

#### **2. Ожидаемые результаты проекта**



Опишите конкретные результаты, которые планируется получить по завершении проекта. Какой программный продукт, методика или информационная система будет создан/создана в результате вашей работы?

По завершении проекта будет создана **веб-ориентированная информационная система семантической сегментации спутниковых снимков Sentinel-2 с возможностью выбора и сравнения нейросетевых моделей**.

Конкретные результаты проекта:

1. **Программный продукт** — веб-приложение, включающее:
   - Клиентскую часть с интерактивной картой на базе Leaflet и панелью управления.
   - Серверную часть на FastAPI, реализующую REST API.
   - Подсистему фоновой обработки задач (Celery + Redis) для асинхронного выполнения ресурсоёмкого инференса нейросетей.
   - Базу данных на PostgreSQL + PostGIS для хранения метаданных и векторных результатов.

2. **Дообученные нейросетевые модели** — адаптированные под данные выбранного региона версии трёх архитектур: **AnySat**, **MF-Mamba**, **CFFormer**. Модели будут дообучены на локальном датасете, включающем снимки Sentinel-2 и эталонные маски сегментации.

3. **Методика сравнительного анализа** — формализованная процедура оценки и сопоставления нейросетевых архитектур по набору метрик (mIoU, F1-Score, время инференса, потребление GPU-памяти), реализованная в виде модуля системы.

4. **Проектная документация** — спецификация требований (SRS), UML-диаграммы, схема базы данных, руководство пользователя, описание API.

5. **Экспериментальные результаты** — количественные и визуальные результаты сравнения трёх моделей на тестовом участке выбранного региона, оформленные в виде таблиц и графиков.

#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

Практическое применение разрабатываемой системы предполагает следующие основные сценарии взаимодействия пользователя с решением.

#### Сценарий 1. «Быстрая сегментация выбранной территории»

1. Пользователь открывает веб-интерфейс системы.
2. На интерактивной карте выбирает область интереса (прямоугольником) либо загружает собственный файл в формате GeoTIFF.
3. В панели управления выбирает задачу сегментации (например, «Водные объекты», «Застройка», «Все классы»).
4. Выбирает одну из доступных нейросетевых моделей из выпадающего списка.
5. Нажимает кнопку «Запустить анализ».
6. Система отображает индикатор выполнения. После завершения обработки на карту накладывается цветной слой сегментации.
7. Пользователь может настроить прозрачность слоя, включить/выключить отображение легенды.
8. Результат можно скачать в формате GeoJSON или Shapefile для дальнейшей работы в QGIS/ArcGIS.

#### Сценарий 2. «Сравнение нескольких моделей»

1. Пользователь выбирает область интереса.
2. В панели управления выбирает опцию «Сравнить все доступные модели».
3. Система последовательно (или параллельно) выполняет инференс всеми тремя моделями.
4. После завершения пользователю доступны:
   - Визуальное сравнение — возможность переключаться между слоями или просматривать их рядом (режим side-by-side).
   - Таблица сравнения — сводка метрик, полученных на тестовом участке (mIoU, F1, время обработки, потребление памяти).
   - Рекомендация — система подсвечивает модель, оптимальную по выбранному критерию (точность/скорость).
5. Пользователь скачивает результат от выбранной модели.

#### Сценарий 3. «Пакетная обработка (продвинутый)»

1. Пользователь загружает несколько GeoTIFF-файлов или указывает временной диапазон для автоматической загрузки снимков Sentinel-2 через Google Earth Engine.
2. Выбирает модель и параметры обработки.
3. Система запускает пакетную обработку в фоновом режиме.
4. Пользователь получает уведомление (email или в интерфейсе) о завершении.
5. Результаты доступны в виде архива векторных слоёв и сводного отчёта.

#### Сценарий 4. «Интеграция через API»

1. Сторонний сервис (например, корпоративная ГИС или мобильное приложение) отправляет POST-запрос к эндпоинту `/api/v1/segment` с координатами области или загруженным файлом.
2. Система возвращает `task_id`.
3. Сторонний сервис периодически опрашивает `/api/v1/task/{task_id}` до получения статуса `completed`.
4. После завершения сервис получает GeoJSON с результатами сегментации и встраивает их в собственный интерфейс.

- Сокращение времени на сегментацию снимка 10×10 км: с **8 часов** (ручная оцифровка) до **2 минут** (автоматическая обработка).

- Возможность обработки до **100 км²** спутниковых данных в час на одном GPU-сервере, что делает систему пригодной для задач регионального масштаба.
